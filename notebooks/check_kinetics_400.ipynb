```python
import pandas as pd

csv_train = "/checkpoint/abardes/datasets/Kinetics400/annotations/k400_train_paths.csv"
csv_val = "/checkpoint/abardes/datasets/Kinetics400/annotations/k400_val_paths.csv"
txt_labels = "/checkpoint/abardes/datasets/Kinetics400/annotations/k400_labels.txt"

df_train = pd.read_csv(csv_train, header=None, names=["video_path", "index"], sep=" ")
df_val = pd.read_csv(csv_val, header=None, names=["video_path", "index"], sep=" ")

import pandas as pd

# Read and split each line
with open(txt_labels, "r") as f:
    lines = [line.strip().split(" ", 1) for line in f]  # split only on first space

# Build dataframe
df_labels = pd.DataFrame(lines, columns=["index", "label"])
df_labels["index"] = df_labels["index"].astype("int64")  # convert to int64
```

```python
df_train = pd.merge(df_train, df_labels, on="index")
df_val = pd.merge(df_val, df_labels, on="index")
```
